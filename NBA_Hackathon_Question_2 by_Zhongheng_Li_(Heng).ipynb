### Convert this answer into python functions for future questions:

```python
def getProb_In4( input_p_H, input_p_A , winOrLose_str ):
     
    if(winOrLose_str == 'lose'):
        p_H = (1 - input_p_H)
        p_A = (1 - input_p_A)
    else: 
        p_H = input_p_H
        p_A = input_p_A
        
    return (p_H**2 ) * (p_A**2 )
```

```python
def getProb_In5( p_H, p_A , winOrLose_str ):
     
    result = 0
    
    if(winOrLose_str == 'lose'):
    
        result = 2 * (p_H * (1 - p_A)**2  *  (1 - p_H)**2) + 2 * ((1 - p_H)**3 * p_A  *  (1 - p_A)) 
        

    else: 
    
       result = 2 * ((1 - p_H) * p_H**2 * p_A**2) +  2 * (p_H**3 * (1 - p_A) * p_A)  
    
    return result

```

```python
## Function to caculate the probability for each team from given conference to get to R4, the NBA Winning 
def getProbsToFinals(conference):
    
    
    for seed in Seeds:    

        thisSeedStr = conference + str(seed)

        TeamWinningPercentagePerRound[thisSeedStr] = {}

        totalWinProb = 0

        for r in Rounds:

            ## Calculate Winning Percentage for this round
            OpponentSeeds= roundMatchesDict[seed][r]
            OpponentSeedsLen = len(OpponentSeeds)

            round_winningProb = 0;

            for OpponentSeed in  OpponentSeeds:

                OpponentSeedStr = conference + str(OpponentSeed)

                if(seed > OpponentSeed):

                    probRow = win_probabilities_df.loc[(win_probabilities_df['Team1'] == OpponentSeedStr) & (win_probabilities_df['Team2'] == thisSeedStr)]

                    p_H = 1 - probRow.Prob1WinsHome
                    p_A = 1 - probRow.Prob1WinsAway


                else:

                    probRow = win_probabilities_df.loc[(win_probabilities_df['Team1'] == thisSeedStr) & (win_probabilities_df['Team2'] == OpponentSeedStr)]

                    p_H = probRow.Prob1WinsHome
                    p_A = probRow.Prob1WinsAway



                round_winningProb  = float(round_winningProb) +  getTotalProb_WinSeries ( p_H,p_A)


            AvgRound_winningProb =  float(round(round_winningProb / OpponentSeedsLen,5)) 

            if (r == 'R1'):
                totalWinProb = AvgRound_winningProb
            else:
                totalWinProb = totalWinProb * AvgRound_winningProb


            TeamWinningPercentagePerRound[thisSeedStr][r] = round(totalWinProb,5)
      
 
```

```python
## Function to get Revenue with given HomeTeam, AwayTeam, the round number and the number of games are playing 
def getRevenue(HomeTeam,AwayTeam, roundNumber, gameNumber):
    HomeTeamRev = Hypothetical_Playoff_Gate_Data_df.loc[(Hypothetical_Playoff_Gate_Data_df['HomeTeam'] == HomeTeam) ]
    AwayTeamRev = Hypothetical_Playoff_Gate_Data_df.loc[(Hypothetical_Playoff_Gate_Data_df['HomeTeam'] == AwayTeam) ]

    RoundX_GateStr = 'RoundX_Gate'

    revenue = 0
    if (gameNumber == 4):

        revenue =  float((HomeTeamRev[RoundX_GateStr.replace('X', str(roundNumber) )])*2) + float((AwayTeamRev[RoundX_GateStr.replace('X', str(roundNumber) )])*2)

    elif (gameNumber == 5):   

         revenue =  float((HomeTeamRev[RoundX_GateStr.replace('X', str(roundNumber) )])*3) + float((AwayTeamRev[RoundX_GateStr.replace('X', str(roundNumber) )])*2)

    elif (gameNumber == 6):   

         revenue =  float((HomeTeamRev[RoundX_GateStr.replace('X', str(roundNumber) )])*3) + float((AwayTeamRev[RoundX_GateStr.replace('X', str(roundNumber) )])*3)

    elif (gameNumber == 7):   

         revenue =  float((HomeTeamRev[RoundX_GateStr.replace('X', str(roundNumber) )])*4) + float((AwayTeamRev[RoundX_GateStr.replace('X', str(roundNumber) )])*3)

   
    #return '${:,.2f}'.format(revenue)
    return revenue
```

```python
## Function to get Revenue for my team with given indication of my team is being a home team or away team, the round number and the number of games are playing 
def getMyTeamRevenue(MyTeam, homeOrAwayStr, roundNumber, gameNumber):
    
    
    MyTeam = Hypothetical_Playoff_Gate_Data_df.loc[(Hypothetical_Playoff_Gate_Data_df['HomeTeam'] == MyTeam) ]
  
    RoundX_GateStr = 'RoundX_Gate'

    revenue = 0
    if (gameNumber == 4):
        
        revenue =  float((MyTeam[RoundX_GateStr.replace('X', str(roundNumber) )])*2) 

    elif (gameNumber == 5):   
        
        if(homeOrAwayStr == 'home'):
                
            revenue =  float((MyTeam[RoundX_GateStr.replace('X', str(roundNumber) )])*3) 

        else:
            revenue =  float((MyTeam[RoundX_GateStr.replace('X', str(roundNumber) )])*2) 


    elif (gameNumber == 6):   

         revenue =  float((MyTeam[RoundX_GateStr.replace('X', str(roundNumber) )])*3) 

    elif (gameNumber == 7):   

        if(homeOrAwayStr == 'home'):
                
            revenue =  float((MyTeam[RoundX_GateStr.replace('X', str(roundNumber) )])*4) 

        else:
            revenue =  float((MyTeam[RoundX_GateStr.replace('X', str(roundNumber) )])*3) 


    #return '${:,.2f}'.format(revenue)
    return revenue
```

```python
## Function to create Playoff Simulator dataframe with given team conference and seed. This function is leverging the outcomes from the previously created functions
def createPlayoffSimulatorDF(conference, seed):

    if(conference == 'East'):
        OppositConference = 'West'
    elif(conference == 'West'):
        OppositConference = 'East'

    df = pd.DataFrame()

    thisSeedStr = conference + str(seed)
    roundNumber = 1

    for r in Rounds:

            OpponentSeeds= roundMatchesDict[seed][r]
            OpponentSeedsLen = len(OpponentSeeds)

            HomeTeam = ''

            AwayTeam = ''
            
            myTeamStatus = ''

            round_winningProb = 0;

            for OpponentSeed in  OpponentSeeds:

                OpponentSeedStr = conference + str(OpponentSeed)

                p_H = 0
                p_A = 0

                if(seed > OpponentSeed):

                    HomeTeam = OpponentSeedStr
                    AwayTeam = thisSeedStr
                    myTeamStatus = 'Away'

                    probRow = win_probabilities_df.loc[(win_probabilities_df['Team1'] == OpponentSeedStr) & (win_probabilities_df['Team2'] == thisSeedStr)]

                    p_H = 1 - probRow.Prob1WinsHome
                    p_A = 1 - probRow.Prob1WinsAway


                else:

                    HomeTeam = thisSeedStr
                    AwayTeam = OpponentSeedStr
                    myTeamStatus = 'Home'

                    probRow = win_probabilities_df.loc[(win_probabilities_df['Team1'] == thisSeedStr) & (win_probabilities_df['Team2'] == OpponentSeedStr)]

                    p_H = probRow.Prob1WinsHome
                    p_A = probRow.Prob1WinsAway


                round_winningProb  =  getTotalProb_WinSeries (p_H,p_A)


                probOfEncounter = 0

                if(r == 'R1'):
                    probOfEncounter = 1
                else:
                    probOfEncounter = TeamWinningPercentagePerRound[OpponentSeedStr][r]

                probOfPlayingGame4 = getProb_In4( p_H,p_A,'win') + getProb_In4(p_H,p_A,'lose') 
                probOfPlayingGame5 = getProb_In5( p_H,p_A,'win') + getProb_In4(p_H,p_A,'lose')
                probOfPlayingGame6 = getProb_In6( p_H,p_A,'win') + getProb_In4(p_H,p_A,'lose')
                probOfPlayingGame7 = getProb_In7( p_H,p_A,'win') + getProb_In4(p_H,p_A,'lose')


                revOfPlayingGame4To7 = []

                TotalRevOfPlayingGame4To7 = []
                homeRevOfPlayingGame4To7 = []

                for i in range(4, 8):

                    TotalRevOfPlayingGame4To7.append(getRevenue(HomeTeam, AwayTeam, roundNumber, i)) 
                    homeRevOfPlayingGame4To7.append(getMyTeamRevenue(thisSeedStr, myTeamStatus, roundNumber, i)) 


    
                df = df.append({'Team': OpponentSeedStr, 
                                        'Round': r, 
                                        'ProbOfEncounter': float(round(probOfEncounter,5)),
                                        'ProbOfWin': float(round(round_winningProb,5)), 
                                        'ProbOfLose': float(round(1 - round_winningProb,5)), 
                                        'Avg_MyTeamRev': '${:,.2f}'.format(((homeRevOfPlayingGame4To7[0] + homeRevOfPlayingGame4To7[3]) / 2)),
                                        'Avg_TotalRev':  '${:,.2f}'.format(((TotalRevOfPlayingGame4To7[0] + TotalRevOfPlayingGame4To7[3]) / 2)), 
                                        'ProbOfPlaying_4G': float(round(probOfPlayingGame4,5)),
                                        'MyTeamRevOfPlaying_4G': '${:,.2f}'.format(homeRevOfPlayingGame4To7[0]),
                                        'TotalRevOfPlaying_4G': '${:,.2f}'.format(TotalRevOfPlayingGame4To7[0]),
                                        'ProbOfPlaying_5G': float(round(probOfPlayingGame5,5)),
                                        'MyTeamRevOfPlaying_5G': '${:,.2f}'.format(homeRevOfPlayingGame4To7[1]),
                                        'TotalRevOfPlaying_5G': '${:,.2f}'.format(TotalRevOfPlayingGame4To7[1]),
                                        'ProbOfPlaying_6G': float(round(probOfPlayingGame6,5)),
                                        'MyTeamRevOfPlaying_6G': '${:,.2f}'.format(homeRevOfPlayingGame4To7[2]),
                                        'TotalRevOfPlaying_6G': '${:,.2f}'.format(TotalRevOfPlayingGame4To7[2]),
                                        'ProbOfPlaying_7G': float(round(probOfPlayingGame7,5)),
                                        'MyTeamRevOfPlaying_7G': '${:,.2f}'.format(homeRevOfPlayingGame4To7[3]),
                                        'TotalRevOfPlaying_7G': '${:,.2f}'.format(TotalRevOfPlayingGame4To7[3])
                                        }, ignore_index=True)


            roundNumber = roundNumber + 1




    for OpponentSeed in Seeds:

        OpponentSeedStr =  OppositConference + str(OpponentSeed)

        HomeTeam = ''

        AwayTeam = ''
        
        myTeamStatus = ''

        round_winningProb = 0;

        p_H = 0
        p_A = 0

        if(seed > OpponentSeed or OppositConference == 'East'):

            
            HomeTeam = OpponentSeedStr
            AwayTeam = thisSeedStr
            
            if(conference != 'West'):
                myTeamStatus = 'Away'
                probRow = win_probabilities_df.loc[(win_probabilities_df['Team1'] == thisSeedStr) & (win_probabilities_df['Team2'] == OpponentSeedStr )]

                p_H = probRow.Prob1WinsHome
                p_A = probRow.Prob1WinsAway
                
            else:
                myTeamStatus = 'Home'
                probRow = win_probabilities_df.loc[(win_probabilities_df['Team1'] == OpponentSeedStr) & (win_probabilities_df['Team2'] == thisSeedStr)]

                p_H = 1 - probRow.Prob1WinsHome
                p_A = 1 - probRow.Prob1WinsAway
           


        elif(seed < OpponentSeed or seed == OpponentSeed ):
            

            HomeTeam = thisSeedStr
            AwayTeam = OpponentSeedStr
            
                     
            if(conference != 'West'):
                myTeamStatus = 'Away'
            else:
                myTeamStatus = 'Home'
       

            probRow = win_probabilities_df.loc[(win_probabilities_df['Team1'] == thisSeedStr) & (win_probabilities_df['Team2'] == OpponentSeedStr)]
            

            p_H = probRow.Prob1WinsHome
            p_A = probRow.Prob1WinsAway


        round_winningProb  =  getTotalProb_WinSeries (p_H,p_A)


        #probOfEncounter = The probability of Opponent Seed won all 3 rounds in their own conferences and got into the finals * The probability of my team won all 3 rounds in our own conferences and got into the finals 
        probOfEncounter = TeamWinningPercentagePerRound[thisSeedStr]['R3'] * TeamWinningPercentagePerRound[OpponentSeedStr]['R3']

        probOfPlayingGame4 = getProb_In4( p_H,p_A,'win') + getProb_In4(p_H,p_A,'lose') 
        probOfPlayingGame5 = getProb_In5( p_H,p_A,'win') + getProb_In4(p_H,p_A,'lose')
        probOfPlayingGame6 = getProb_In6( p_H,p_A,'win') + getProb_In4(p_H,p_A,'lose')
        probOfPlayingGame7 = getProb_In7( p_H,p_A,'win') + getProb_In4(p_H,p_A,'lose')


        TotalRevOfPlayingGame4To7 = []
        homeRevOfPlayingGame4To7 = []

        for i in range(4, 8):

            TotalRevOfPlayingGame4To7.append(getRevenue(HomeTeam, AwayTeam, roundNumber, i)) 
            homeRevOfPlayingGame4To7.append(getMyTeamRevenue(thisSeedStr, myTeamStatus, roundNumber, i)) 

  
            

        df = df.append({'Team': OpponentSeedStr, 
                                'Round': 'R4', 
                                'ProbOfEncounter': float(round(probOfEncounter,5)),
                                'ProbOfWin': float(round(round_winningProb,5)), 
                                'ProbOfLose': float(round(1 - round_winningProb,5)), 
                                'Avg_MyTeamRev': '${:,.2f}'.format(((homeRevOfPlayingGame4To7[0] + homeRevOfPlayingGame4To7[3]) / 2)),
                                'Avg_TotalRev':  '${:,.2f}'.format(((TotalRevOfPlayingGame4To7[0] + TotalRevOfPlayingGame4To7[3]) / 2)), 
                                'ProbOfPlaying_4G': float(round(probOfPlayingGame4,5)),
                                'MyTeamRevOfPlaying_4G': '${:,.2f}'.format(homeRevOfPlayingGame4To7[0]),
                                'TotalRevOfPlaying_4G': '${:,.2f}'.format(TotalRevOfPlayingGame4To7[0]),
                                'ProbOfPlaying_5G': float(round(probOfPlayingGame5,5)),
                                'MyTeamRevOfPlaying_5G': '${:,.2f}'.format(homeRevOfPlayingGame4To7[1]),
                                'TotalRevOfPlaying_5G': '${:,.2f}'.format(TotalRevOfPlayingGame4To7[1]),
                                'ProbOfPlaying_6G': float(round(probOfPlayingGame6,5)),
                                'MyTeamRevOfPlaying_6G': '${:,.2f}'.format(homeRevOfPlayingGame4To7[2]),
                                'TotalRevOfPlaying_6G': '${:,.2f}'.format(TotalRevOfPlayingGame4To7[2]),
                                'ProbOfPlaying_7G': float(round(probOfPlayingGame7,5)),
                                'MyTeamRevOfPlaying_7G': '${:,.2f}'.format(homeRevOfPlayingGame4To7[3]),
                                'TotalRevOfPlaying_7G': '${:,.2f}'.format(TotalRevOfPlayingGame4To7[3])
                                }, ignore_index=True)


        
      
     
    cols = ['Team' , 'Round','ProbOfEncounter', 'ProbOfWin', 'ProbOfLose', 'Avg_MyTeamRev', 'Avg_TotalRev', 'ProbOfPlaying_4G', 'MyTeamRevOfPlaying_4G', 'TotalRevOfPlaying_4G', 'ProbOfPlaying_5G','MyTeamRevOfPlaying_5G', 'TotalRevOfPlaying_5G', 'ProbOfPlaying_6G', 'MyTeamRevOfPlaying_6G', 'TotalRevOfPlaying_6G','ProbOfPlaying_7G', 'MyTeamRevOfPlaying_7G','TotalRevOfPlaying_7G']
    df = df[cols]
        

    return df
```

```python
#function to get expected revenues from own conference
def getConferenceCumulativeRev(conference,seed,playoffSimulatorDF):

    thisSeedStr = conference + str(seed)
    
    df = pd.DataFrame()
    
    
    cumulativeMyTeamRev = 0
    cumulativeTotalRev = 0
    
    for i in range(0,7):

        cumulativeMyTeamRev = float(cumulativeMyTeamRev) + playoffSimulatorDF.loc[[i]]['ProbOfEncounter']  * playoffSimulatorDF.loc[[i]]['Avg_MyTeamRev'].replace( '[\$,)]','', regex=True).astype(float)
        cumulativeTotalRev = float(cumulativeTotalRev) + playoffSimulatorDF.loc[[i]]['ProbOfEncounter']  * playoffSimulatorDF.loc[[i]]['Avg_TotalRev'].replace( '[\$,)]','', regex=True).astype(float)
    
    
    
    df = df.append({'Conference': conference, 
                            'Seed': seed, 
                            'ProbOfEarningThisRevenue': TeamWinningPercentagePerRound[thisSeedStr]['R3'],
                            'Cumulative_MyTeamRev': '${:,.2f}'.format(float(cumulativeMyTeamRev)), 
                            'Cumulative_TotalRev': '${:,.2f}'.format(float(cumulativeTotalRev))    
                            }, ignore_index=True)

     
    cols = ['Conference' , 'Seed','ProbOfEarningThisRevenue', 'Cumulative_MyTeamRev', 'Cumulative_TotalRev']
    df = df[cols]
        

    return df
```

```python
## function to get cumuated revenue after finals with given my team conference, my team seed, opponent team conference and opponent team seed
def getCumulatedRevenueAfterFinals(myTeamConference,myTeamSeed,OpponentTeamConference,OpponentTeamSeed):
    df = pd.DataFrame()

    thisSeedStr = myTeamConference + str(myTeamSeed)
    OpponentSeedStr = OpponentTeamConference + str(OpponentTeamSeed)



    myTeamPlayoffSimulator_DF = createPlayoffSimulatorDF(myTeamConference, myTeamSeed)

    opponentTeamPlayoffSimulator_DF = createPlayoffSimulatorDF(OpponentTeamConference, OpponentTeamSeed)


    myTeamConference_DF = getConferenceCumulativeRev(myTeamConference,myTeamSeed,myTeamPlayoffSimulator_DF)

    opponentTeamConference_DF = getConferenceCumulativeRev(OpponentTeamConference,OpponentTeamSeed,opponentTeamPlayoffSimulator_DF)


    cumulative_MyTeamRev = float(myTeamConference_DF.Cumulative_MyTeamRev.replace( '[\$,)]','', regex=True).astype(float)) + float(myTeamPlayoffSimulator_DF.loc[(myTeamPlayoffSimulator_DF['Team'] == OpponentSeedStr)].Avg_MyTeamRev.replace( '[\$,)]','', regex=True).astype(float))

    cumulative_TotalRev =  float(opponentTeamConference_DF.Cumulative_TotalRev.replace( '[\$,)]','', regex=True).astype(float))+ float(myTeamConference_DF.Cumulative_TotalRev.replace( '[\$,)]','', regex=True).astype(float)) + float(myTeamPlayoffSimulator_DF.loc[(myTeamPlayoffSimulator_DF['Team'] == OpponentSeedStr)].Avg_TotalRev.replace( '[\$,)]','', regex=True).astype(float))


    df = df.append({'MyTeam': thisSeedStr, 
                        'OpponentTeam': OpponentSeedStr, 
                        'Cumulative_MyTeamRev': '${:,.2f}'.format(float(cumulative_MyTeamRev)), 
                        'Cumulative_BothTeamsTotalRev': '${:,.2f}'.format(float(cumulative_TotalRev))
                        }, ignore_index=True)


    cols = ['MyTeam' , 'OpponentTeam', 'Cumulative_MyTeamRev', 'Cumulative_BothTeamsTotalRev']

    df = df[cols]


    return df
```
